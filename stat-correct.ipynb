# Analysis of 'correct' by Group

This notebook loads the dataset `Milestones_survival_Final.csv`, filters it to include only the `control` and `treatment` groups, creates a contingency table of the `correct` variable, and computes Fisher's exact test and the chi-square test.

```python
import pandas as pd
import scipy.stats as stats

# Load the dataset
#df = pd.read_csv('./data/Milestones_survival_Final.csv')
df = pd.read_csv('./data/Milestones_survival_Final_dbscam.csv')

print('Dataset loaded successfully.')
df.head()
```

```python
# Filter data to include only 'control' and 'treatment' groups
df_filtered = df[df['group'].isin(['Control', 'Treatment'])]

# Create a contingency table for 'group' vs 'correct'
contingency_table = pd.crosstab(df_filtered['group'], df_filtered['correct'])
print('Contingency Table:')
print(contingency_table)
```

```python
# Perform Fisher's Exact Test (suitable for a 2x2 contingency table)
fisher_oddsratio, fisher_p = stats.fisher_exact(contingency_table)

print("Fisher's Exact Test Results:")
print(f"Odds Ratio: {fisher_oddsratio}")
print(f"P-value: {fisher_p}")
```

```python
# Perform Chi-Square Test
chi2, chi2_p, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-Square Test Results:")
print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {chi2_p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```
